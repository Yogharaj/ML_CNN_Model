```python
import tensorflow as tf
import os
from tensorflow.keras.models import load_model
import cv2
import imghdr
import matplotlib.pyplot as plt
import numpy as np
mod=load_model("C:\\Users\\muhil\\Documents\\Machine_learning_project\\model\\DogCat.h5")
img=cv2.imread("C:\\Users\\muhil\\OneDrive\\Pictures\\Dog (30).jpg")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
image=tf.image.resize(img,(256,256))
y_pred=mod.predict(np.expand_dims(image/255, 0))
print(y_pred)
if y_pred>0.5:
    print(f'Predicted class is Dog')
else:
    print(f'Predicted class is Cat')
```
